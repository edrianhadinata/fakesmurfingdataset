```python
import csv, os
import pandas as pd
from datetime import datetime, timedelta
```

```python
df_edge = pd.read_csv('dataset_edges.csv')
df_edge_list = list(df_edge.itertuples(index=False, name=None))
```

```python
edges_list = [(u, v) for (u, v, w, t) in df_edge_list]
#print(edges_list)
```

```python
import networkx as nx
```

```python
G_all = nx.DiGraph()
G_all.add_edges_from(edges_list)
```

```python
# Layout (spring_layout atau yang lain)
edge_lay = nx.spring_layout(G_all, seed=42, k=2)
width = 0.2
```

```python
import matplotlib.pyplot as plt
```

```python
# --------- FIGURE: PLOT ---------
plt.figure(figsize=(12, 9))
nx.draw(G_all, edge_lay, with_labels=True, font_size=7, width=width, node_size=200, linewidths=0.4, edgecolors='black', node_color='lightgreen', arrows=True)
plt.title(f"Graf Hasil Data Fake Transaksi")
plt.show()
```
